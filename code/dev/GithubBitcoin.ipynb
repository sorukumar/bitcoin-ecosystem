```python
repositories = {
  "bitcoin/bitcoin": "bitcoin"
}

# Define columns for current data
current_columns = [
  'Repository', 'Short Name', 'Commits', 'Open Issues', 'Total Issues', 'Open PRs', 'Total PRs', 
  'Contributors', 'Authors', 'Committers', 'Forks', 'Stars', 'Watchers', 'Languages'
]

def get_contributor_count(repo_name):
  url = f"https://api.github.com/repos/{repo_name}/contributors?per_page=1&anon=true"
  headers = {'Authorization': f'token {ACCESS_TOKEN}'}
  response = requests.get(url, headers=headers)
  if response.status_code == 200:
      links = response.links
      if 'last' in links:
          last_url = links['last']['url']
          return int(last_url.split('page=')[-1])
  return None

def get_language_percentages(repo):
  languages = repo.get_languages()
  if not languages:
      return "No languages detected"
  total = sum(languages.values())
  return ', '.join([f"{lang} ({count/total*100:.1f}%)" for lang, count in languages.items()])

def get_issue_count(repo, state):
  return repo.get_issues(state=state).totalCount - repo.get_pulls(state=state).totalCount

def handle_rate_limit():
  rate_limit = g.get_rate_limit()
  reset_time = rate_limit.core.reset.timestamp()
  sleep_time = reset_time - time.time() + 1  # Add 1 second buffer
  print(f"Rate limit exceeded. Sleeping for {sleep_time:.2f} seconds.")
  time.sleep(sleep_time)

def get_author_count(repo):
  authors = set()
  for commit in repo.get_commits():
      try:
          authors.add(commit.commit.author.name)
      except RateLimitExceededException:
          handle_rate_limit()
          authors.add(commit.commit.author.name)
  return len(authors)

def get_committer_count(repo):
  committers = set()
  for commit in repo.get_commits():
      try:
          committers.add(commit.commit.committer.name)
      except RateLimitExceededException:
          handle_rate_limit()
          committers.add(commit.commit.committer.name)
  return len(committers)

def get_current_repo_data(repositories):
  data_list = []
  for repo_name, short_name in repositories.items():
      print(f"Processing current data for {repo_name}")
      while True:
          try:
              repo = g.get_repo(repo_name)
              
              # Fetch current data
              data = {
                  'Repository': repo_name,
                  'Short Name': short_name,
                  'Commits': repo.get_commits().totalCount,
                  'Open Issues': get_issue_count(repo, 'open'),
                  'Total Issues': get_issue_count(repo, 'all'),
                  'Open PRs': repo.get_pulls(state='open').totalCount,
                  'Total PRs': repo.get_pulls(state='all').totalCount,
                  'Contributors': get_contributor_count(repo_name),
                  'Authors': get_author_count(repo),
                  'Committers': get_committer_count(repo),
                  'Forks': repo.forks_count,
                  'Stars': repo.stargazers_count,
                  'Watchers': repo.get_subscribers().totalCount,
                  'Languages': get_language_percentages(repo)
              }
              data_list.append(data)
              break  # Exit the while loop if successful
          except RateLimitExceededException:
              handle_rate_limit()
          except Exception as e:
              print(f"Failed to process {repo_name}: {e}")
              break  # Exit the while loop if there's a non-rate-limit error

  return pd.DataFrame(data_list, columns=current_columns)

# Main execution
if __name__ == "__main__":
  current_repo_data = get_current_repo_data(repositories)
```

```python


def handle_rate_limit(response):
  if 'X-RateLimit-Remaining' in response.headers and int(response.headers['X-RateLimit-Remaining']) == 0:
      reset_time = int(response.headers['X-RateLimit-Reset'])
      sleep_time = reset_time - time.time() + 1
      print(f"Rate limit exceeded. Sleeping for {sleep_time:.2f} seconds.")
      time.sleep(sleep_time)

def get_all_contributors(repo_name):
  url = f"https://api.github.com/repos/{repo_name}/contributors"
  params = {"per_page": 100, "anon": "true"}
  headers = {"Authorization": f"token {ACCESS_TOKEN}"}
  contributors = []
  page = 1

  while True:
      response = requests.get(url, params=params, headers=headers)
      handle_rate_limit(response)
      
      if response.status_code != 200:
          print(f"Error fetching contributors: {response.status_code}")
          break

      page_contributors = response.json()
      if not page_contributors:
          break

      contributors.extend(page_contributors)
      print(f"Fetched page {page} of contributors. Total: {len(contributors)}")
      
      if 'next' not in response.links:
          break
      
      url = response.links['next']['url']
      page += 1

  return contributors

def get_contributor_details(repo_name, contributor):
  headers = {"Authorization": f"token {ACCESS_TOKEN}"}
  
  if contributor.get('type') == 'Anonymous':
      return {
          'Login': 'Anonymous',
          'Contributions': contributor.get('contributions', 0),
          'Name': contributor.get('name'),
          'Email': contributor.get('email'),
          'First Contribution': None,
          'Last Contribution': None,
      }

  login = contributor.get('login')
  if not login:
      print(f"Warning: Unexpected contributor format: {contributor}")
      return None

  user_url = f"https://api.github.com/users/{login}"
  user_response = requests.get(user_url, headers=headers)
  handle_rate_limit(user_response)
  
  user_details = user_response.json() if user_response.status_code == 200 else {}

  commits_url = f"https://api.github.com/repos/{repo_name}/commits"
  params = {"author": login, "per_page": 100}
  commits = []
  while True:
      commits_response = requests.get(commits_url, params=params, headers=headers)
      handle_rate_limit(commits_response)
      if commits_response.status_code != 200:
          break
      page_commits = commits_response.json()
      if not page_commits:
          break
      commits.extend(page_commits)
      if 'next' not in commits_response.links:
          break
      commits_url = commits_response.links['next']['url']

  if commits:
      first_commit = datetime.strptime(commits[-1]['commit']['author']['date'], "%Y-%m-%dT%H:%M:%SZ").replace(tzinfo=timezone.utc)
      last_commit = datetime.strptime(commits[0]['commit']['author']['date'], "%Y-%m-%dT%H:%M:%SZ").replace(tzinfo=timezone.utc)
      now = datetime.now(timezone.utc)
      
      contribution_duration = (last_commit - first_commit).days
      recent_commits = sum(1 for commit in commits if (now - datetime.strptime(commit['commit']['author']['date'], "%Y-%m-%dT%H:%M:%SZ").replace(tzinfo=timezone.utc)).days <= 90)
      avg_commits_per_month = len(commits) / (contribution_duration / 30) if contribution_duration > 0 else 0
  else:
      first_commit = last_commit = None
      contribution_duration = recent_commits = avg_commits_per_month = 0

  return {
      'Login': login,
      'Contributions': contributor['contributions'],
      'URL': contributor.get('html_url'),
      'ID': contributor['id'],
      'First Contribution': first_commit,
      'Last Contribution': last_commit,
      'Contribution Duration (days)': contribution_duration,
      'Recent Contributions (90 days)': recent_commits,
      'Avg Commits per Month': avg_commits_per_month,
      'Name': user_details.get('name'),
      'Email': user_details.get('email'),
      'Company': user_details.get('company'),
      'Location': user_details.get('location'),
      'Public Repos': user_details.get('public_repos'),
      'Followers': user_details.get('followers'),
      'Created At': user_details.get('created_at'),
  }

def assign_percentile(rank, total):
  percentile = rank / total * 100
  if percentile <= 1:
      return 'Top 1%'
  elif percentile <= 5:
      return 'Top 5%'
  elif percentile <= 20:
      return 'Top 20%'
  elif percentile <= 40:
      return 'Top 40%'
  elif percentile <= 60:
      return 'Top 60%'
  elif percentile <= 80:
      return 'Top 80%'    
  else:
      return 'Bottom 20%'

def get_full_contributor_details(repositories):
  all_data = []
  for repo_name in repositories:
      print(f"Processing contributors for {repo_name}")
      contributors = get_all_contributors(repo_name)
      print(f"Total contributors found: {len(contributors)}")
      for i, contributor in enumerate(contributors, 1):
          details = get_contributor_details(repo_name, contributor)
          if details:
              details['Repository'] = repo_name
              all_data.append(details)
          else:
              print(f"Skipped contributor {i}/{len(contributors)}: Unable to process")

  df = pd.DataFrame(all_data)
  df['First Contribution'] = pd.to_datetime(df['First Contribution'], errors='coerce')
  df['Last Contribution'] = pd.to_datetime(df['Last Contribution'], errors='coerce')
  df['First Contribution Month'] = df['First Contribution'].dt.strftime('%Y-%m')
  df['Last Contribution Month'] = df['Last Contribution'].dt.strftime('%Y-%m')
  df['First_Contribution_Year'] = df['First Contribution'].dt.year
  df['Total_Repo_Contributions'] = df.groupby('Repository')['Contributions'].transform('sum')
  df['Contribution_Percentage'] = (df['Contributions'] / df['Total_Repo_Contributions']).round(4)

  df['Contributor_Decile'] = df.groupby('Repository')['Contributions'].transform(
      lambda x: pd.qcut(x, q=10, labels=False, duplicates='drop')
  ) + 1

  sorted_df = df.sort_values('Contributions', ascending=False)
  total_contributors = len(sorted_df)
  df['Contributor_Percentile'] = sorted_df.reset_index().index.map(lambda x: assign_percentile(x + 1, total_contributors))

  return df

```

```python

def handle_rate_limit(response):
  if 'X-RateLimit-Remaining' in response.headers and int(response.headers['X-RateLimit-Remaining']) == 0:
      reset_time = int(response.headers['X-RateLimit-Reset'])
      sleep_time = reset_time - time.time() + 1
      print(f"Rate limit exceeded. Sleeping for {sleep_time:.2f} seconds.")
      time.sleep(sleep_time)

def get_sample_contributors(repo_name, sample_size=5):
  url = f"https://api.github.com/repos/{repo_name}/contributors"
  params = {"per_page": sample_size, "anon": "true"}
  headers = {"Authorization": f"token {ACCESS_TOKEN}"}
  
  response = requests.get(url, params=params, headers=headers)
  handle_rate_limit(response)
  
  if response.status_code != 200:
      print(f"Error fetching contributors: {response.status_code}")
      return []

  return response.json()

def categorize_merge_commit(commit_data):
  files = commit_data.get('files', [])
  
  categories = {
      'bitcoin_core': 0,
      'wallet': 0,
      'gui': 0,
      'tests': 0,
      'docs': 0,
      'build': 0,
      'net': 0,
      'consensus': 0,
      'rpc': 0,
      'other': 0
  }
  
  total_changes = 0
  for file in files:
      filename = file['filename'].lower()
      total_changes += file.get('additions', 0) + file.get('deletions', 0)
      if 'test' in filename:
          categories['tests'] += 1
      elif 'wallet' in filename:
          categories['wallet'] += 1
      elif 'qt' in filename or 'gui' in filename:
          categories['gui'] += 1
      elif filename.endswith(('.md', '.rst')) or 'doc' in filename:
          categories['docs'] += 1
      elif 'src' in filename and filename.endswith(('.cpp', '.h', '.c')):
          if 'wallet' in filename:
              categories['wallet'] += 1
          elif 'net' in filename:
              categories['net'] += 1
          elif 'consensus' in filename:
              categories['consensus'] += 1
          elif 'rpc' in filename:
              categories['rpc'] += 1
          else:
              categories['bitcoin_core'] += 1
      elif 'build' in filename or filename.endswith(('.sh', '.py', 'Makefile')):
          categories['build'] += 1
      else:
          categories['other'] += 1
  
  if len(files) <= 2 and total_changes < 50:
      categories['size'] = 'Small'
  elif len(files) <= 10 and total_changes < 300:
      categories['size'] = 'Medium'
  else:
      categories['size'] = 'Large'
  
  return categories

def get_contributor_details(repo_name, contributor, commit_sample_size=100):
  headers = {"Authorization": f"token {ACCESS_TOKEN}"}
  
  if contributor.get('type') == 'Anonymous':
      return {
          'Login': 'Anonymous',
          'Contributions': contributor.get('contributions', 0),
          'Name': contributor.get('name'),
          'Email': contributor.get('email'),
          'First Contribution': None,
          'Last Contribution': None,
      }

  login = contributor.get('login')
  if not login:
      print(f"Warning: Unexpected contributor format: {contributor}")
      return None

  user_url = f"https://api.github.com/users/{login}"
  user_response = requests.get(user_url, headers=headers)
  handle_rate_limit(user_response)
  
  user_details = user_response.json() if user_response.status_code == 200 else {}

  commits_url = f"https://api.github.com/repos/{repo_name}/commits"
  params = {"per_page": commit_sample_size, "author": login}
  commits_response = requests.get(commits_url, params=params, headers=headers)
  handle_rate_limit(commits_response)
  
  if commits_response.status_code != 200:
      print(f"Error fetching commits for {repo_name}: {commits_response.status_code}")
      return None

  commits = commits_response.json()
  categories_sum = {category: 0 for category in ['bitcoin_core', 'wallet', 'gui', 'tests', 'docs', 'build', 'net', 'consensus', 'rpc', 'other']}
  size_counts = {'Small': 0, 'Medium': 0, 'Large': 0}
  merge_commits = 0
  total_commits = len(commits)
  contributor_commits = []

  for commit in commits:
      if len(commit['parents']) > 1:  # This is a merge commit
          merge_commits += 1
          commit_sha = commit['sha']
          commit_url = f"https://api.github.com/repos/{repo_name}/commits/{commit_sha}"
          commit_response = requests.get(commit_url, headers=headers)
          handle_rate_limit(commit_response)
          if commit_response.status_code == 200:
              commit_details = commit_response.json()
              categories = categorize_merge_commit(commit_details)
              for category, value in categories.items():
                  if category == 'size':
                      size_counts[value] += 1
                  else:
                      categories_sum[category] += value
      contributor_commits.append(commit)

  if contributor_commits:
      first_commit = datetime.strptime(contributor_commits[-1]['commit']['author']['date'], "%Y-%m-%dT%H:%M:%SZ").replace(tzinfo=timezone.utc)
      last_commit = datetime.strptime(contributor_commits[0]['commit']['author']['date'], "%Y-%m-%dT%H:%M:%SZ").replace(tzinfo=timezone.utc)
      now = datetime.now(timezone.utc)
      
      contribution_duration = (last_commit - first_commit).days
      recent_commits = sum(1 for commit in contributor_commits if (now - datetime.strptime(commit['commit']['author']['date'], "%Y-%m-%dT%H:%M:%SZ").replace(tzinfo=timezone.utc)).days <= 90)
      avg_commits_per_month = len(contributor_commits) / (contribution_duration / 30) if contribution_duration > 0 else 0
  else:
      first_commit = last_commit = None
      contribution_duration = recent_commits = avg_commits_per_month = 0

  return {
      'Login': login,
      'Total Commits': total_commits,
      'Merge Commits': merge_commits,
      'URL': contributor.get('html_url'),
      'ID': contributor['id'],
      'First Contribution': first_commit,
      'Last Contribution': last_commit,
      'Contribution Duration (days)': contribution_duration,
      'Recent Contributions (90 days)': recent_commits,
      'Avg Commits per Month': avg_commits_per_month,
      'Name': user_details.get('name'),
      'Email': user_details.get('email'),
      'Company': user_details.get('company'),
      'Location': user_details.get('location'),
      'Public Repos': user_details.get('public_repos'),
      'Followers': user_details.get('followers'),
      'Created At': user_details.get('created_at'),
      'Bitcoin Core Contributions': categories_sum['bitcoin_core'],
      'Wallet Contributions': categories_sum['wallet'],
      'GUI Contributions': categories_sum['gui'],
      'Test Contributions': categories_sum['tests'],
      'Documentation Contributions': categories_sum['docs'],
      'Build Contributions': categories_sum['build'],
      'Networking Contributions': categories_sum['net'],
      'Consensus Contributions': categories_sum['consensus'],
      'RPC Contributions': categories_sum['rpc'],
      'Other Contributions': categories_sum['other'],
      'Small Merge Commits': size_counts['Small'],
      'Medium Merge Commits': size_counts['Medium'],
      'Large Merge Commits': size_counts['Large'],
  }

def get_sample_contributor_details(repositories, contributor_sample_size=5, commit_sample_size=100):
  all_data = []
  for repo_name in repositories:
      print(f"Processing contributors for {repo_name}")
      contributors = get_sample_contributors(repo_name, contributor_sample_size)
      print(f"Sample contributors found: {len(contributors)}")
      for i, contributor in enumerate(contributors, 1):
          details = get_contributor_details(repo_name, contributor, commit_sample_size)
          if details:
              details['Repository'] = repo_name
              all_data.append(details)
          else:
              print(f"Skipped contributor {i}/{len(contributors)}: Unable to process")

  df = pd.DataFrame(all_data)
  
  for col in ['First Contribution', 'Last Contribution']:
      if col not in df.columns:
          print(f"Warning: '{col}' column is missing from the DataFrame")
          df[col] = None

  df['First Contribution'] = pd.to_datetime(df['First Contribution'], errors='coerce')
  df['Last Contribution'] = pd.to_datetime(df['Last Contribution'], errors='coerce')
  df['First Contribution Month'] = df['First Contribution'].dt.strftime('%Y-%m')
  df['Last Contribution Month'] = df['Last Contribution'].dt.strftime('%Y-%m')
  df['First_Contribution_Year'] = df['First Contribution'].dt.year

  return df

# Use the updated function with sample sizes
repositories = ["bitcoin/bitcoin"]
sample_contributors_data = get_sample_contributor_details(repositories, contributor_sample_size=10, commit_sample_size=200)

# Optionally, save to CSV
sample_contributors_data.to_csv('sample_bitcoin_contributors_analysis.csv', index=False)
```

```python

# Step 1: Create a dataframe for anonymous contributors
anonymous_df = detailed_contributors_data[detailed_contributors_data['Login'] == 'Anonymous'].copy()

# Step 2: Create a function to generate a unique identifier for each group
def get_group_id(row, group_dict):
  email = row['Email']
  name = row['Name']
  if pd.notna(email) and email in group_dict:
      return group_dict[email]
  elif pd.notna(name) and name in group_dict:
      return group_dict[name]
  else:
      new_id = len(group_dict)
      if pd.notna(email):
          group_dict[email] = new_id
      if pd.notna(name):
          group_dict[name] = new_id
      return new_id

# Step 3: Apply the grouping
group_dict = {}
anonymous_df['group_id'] = anonymous_df.apply(lambda row: get_group_id(row, group_dict), axis=1)

# Step 4: Aggregate the data for each group
consolidated_anonymous = anonymous_df.groupby('group_id').agg({
  'Email': lambda x: ', '.join(set(filter(pd.notna, x))),
  'Name': lambda x: ', '.join(set(filter(pd.notna, x))),
  'Contributions': 'sum',
  'Contribution_Percentage': 'sum',
  'First Contribution': 'min',
  'Last Contribution': 'max',
  'First_Contribution_Year': 'min',
  'Repository': lambda x: ', '.join(set(x)),
  'Contributor_Percentile': 'first',
})

# Reset the index
consolidated_anonymous.reset_index(drop=True, inplace=True)

# Add back the 'Login' column
consolidated_anonymous['Login'] = 'Anonymous'

# Step 5: Remove the original anonymous entries and add the consolidated ones
non_anonymous_df = detailed_contributors_data[detailed_contributors_data['Login'] != 'Anonymous']
final_df = pd.concat([non_anonymous_df, consolidated_anonymous], ignore_index=True)

# Sort the dataframe by Contributions in descending order
final_df = final_df.sort_values('Contributions', ascending=False)

# Recalculate the Contribution_Percentage
final_df['Total_Repo_Contributions'] = final_df.groupby('Repository')['Contributions'].transform('sum')
# final_df['Contribution_Percentage'] = (final_df['Contributions'] / final_df['Total_Repo_Contributions']).round(4)

# # Recalculate the Contributor_Percentile
# total_contributors = len(final_df)
# final_df['Contributor_Percentile'] = final_df.reset_index().index.map(lambda x: assign_percentile(x + 1, total_contributors))

# Print the consolidated anonymous contributors
print(final_df[final_df['Login'] == 'Anonymous'][['Login', 'Name', 'Email', 'Contributions', 'Contribution_Percentage']].head(10))

# You can now use final_df for further analysis or save it to a file
```
